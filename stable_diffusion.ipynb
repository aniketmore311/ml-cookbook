```python
# diffusion model
diffusion_model = "XpucT/Deliberate" #@param ["XpucT/Deliberate","SG161222/Realistic_Vision_V1.3","eimiss/EimisAnimeDiffusion_1.0v", "andite/anything-v4.0","stabilityai/stable-diffusion-2-1-base","dreamlike-art/dreamlike-photoreal-2.0","Lykon/DreamShaper","gsdf/Counterfeit-V2.5"]
```

```python
# diffusion pipeline setup
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler, DPMSolverMultistepScheduler
import torch

diffusion_pipeline = StableDiffusionPipeline.from_pretrained(diffusion_model, torch_dtype=torch.float16)
diffusion_pipeline.to("cuda")
```

```python
diffusion_pipeline.scheduler = EulerDiscreteScheduler.from_config(diffusion_pipeline.scheduler.config)
diffusion_pipeline.safety_checker = None
```

```python
# generation parameters
import random
from IPython.display import display

image_prompt = "photo, 50mm, samurai, standing on road, in rain, heavy rain, in city, cyberpunk style, cityscape, futuristic" #@param {type: "string"}
suffix = "cinematic, aesthetic, beautiful, photorealistic, hyperrealistic, extremely detailed, highly detailed, high resolution, intricate, masterpiece, award winning, elegant" #@param {type:"string"}
# for reference
# suffix = "cinematic, aesthetic, beautiful, photorealistic, hyperrealistic, extremely detailed, highly detailed, high resolution, intricate, masterpiece, award winning, elegant" #@param {type:"string"}
negative_prompt = "deformed, distorted, disfigured, poorly drawn, bad anatomy, wrong anatomy, extra limb, missing limb, floating limbs, mutated hands and fingers, disconnected limbs, mutation, mutated, ugly, disgusting, blurry, amputation" #@param {type: "string"}
# for reference
# negative_prompt = "deformed, distorted, disfigured, poorly drawn, bad anatomy, wrong anatomy, extra limb, missing limb, floating limbs, mutated hands and fingers, disconnected limbs, mutation, mutated, ugly, disgusting, blurry, amputation"
height_in_px = 512 #@param {type:"number"}
width_in_px = 512 #@param {type:"number"}
# 512:512 -> 1:2
# 512:896 -> 4:7
# 512:912 -> 9:16
inference_steps = 25 #@param {type:"number"}
no_of_images = 3 #@param {type:"number"}
guidance_scale = 7.5 #@param {type:"number"}

prompt = image_prompt + " " + suffix


for i in range(no_of_images):
  seed = random.randint(1,1000000)
  # seed = 352942009
  generator = torch.manual_seed(seed)
  image = diffusion_pipeline(prompt=prompt, generator=generator, negative_prompt=negative_prompt,height=height_in_px,width=width_in_px, num_inference_steps=inference_steps, guidance_scale=guidance_scale).images[0]
  print("seed: "+str(seed))
  display(image)
  print("\n")
```
